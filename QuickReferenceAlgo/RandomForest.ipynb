```python
###calculating the IQR for all the variables
```

```python
Q1=data.quantile(.25)
Q3=data.quantile(.75)
IQR=Q3-Q1
```

```python
upper_bound = Q3 + 1.5 * IQR
```

```python
lower_bound = Q1 - 1.5 * IQR
```

```python
data_filtered = data[((data >= lower_bound) & (data <= upper_bound)).all(axis=1)]
```

```python
data_filtered.describe()
```

* we have to perform standardization because some maximum values are very high which may cause bias towards that variable.

```python
data_filtered.skew()
```

```python
data_filtered.kurt()
```

```python
from sklearn.preprocessing import RobustScaler
robust_scaler = RobustScaler(quantile_range=(5, 95))
transformed_data = robust_scaler.fit_transform(data_filtered)
```

```python
transformed_data=pd.DataFrame(transformed_data,columns=data_filtered.columns)
```

```python
transformed_data.head()
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
X_train, X_test, y_train, y_test = train_test_split(transformed_data, data.target[:398], test_size=0.2, random_state=42)
```

```python
rf_classifier = RandomForestClassifier()
```

```python
param_grid = {
    'n_estimators': [50, 100, 150],
    'max_depth': [None, 5, 10, 20],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}
```

```python
from sklearn.model_selection import GridSearchCV
grid_search = GridSearchCV(estimator=rf_classifier, param_grid=param_grid, cv=5, scoring='accuracy')
```

```python
grid_search.fit(X_train, y_train)
```

```python
best_params = grid_search.best_params_
```

```python
best_params
```

```python
best_rf_classifier = RandomForestClassifier(**best_params, random_state=42)
best_rf_classifier.fit(X_train, y_train)
```

```python
y_pred = best_rf_classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
```

```python
import seaborn as sns
plt.figure(figsize=(8, 6))
sns.heatmap(cm,fmt='g',xticklabels=data.target_names, yticklabels=data.target_names,cbar=False,annot=True)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```
